### Folium

[Folium](https://github.com/python-visualization/folium) makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

```python
import pandas as pd
import re
import numpy as np
import folium #Allow to visualize things on map
import matplotlib.pyplot as plt
%matplotlib inline
from bs4 import BeautifulSoup
import requests
```

```python
mapit = None
latlon = [ (3.144473, 101.708722), (3.144473, 101.708722),(3.135732, 101.686989)]
for coord in latlon:
    mapit = folium.Map( location=[ coord[0], coord[1] ] )
    folium.Marker([3.144473, 101.708722], icon=folium.Icon(color='blue',icon='star'), popup='Federal Hotel').add_to(mapit)
    folium.Marker([3.156374, 101.714579], icon=folium.Icon(color='green',icon='info-sign') , popup='Mandarin Oriental').add_to(mapit)
    folium.Marker([3.135732, 101.686989], icon=folium.Icon(color='red',icon='star') , popup='Le Meridien').add_to(mapit)
mapit
```

```python
# scrape from lelong

# find the pattern for the first page
url = 'https://www.lelong.com.my/catalog/all/list?TheKeyword=macbook+pro&D='

# write a loop to scrape from page 1 to the last page

product_name=[]
for page in range(1,20):
    url_page = url+str(page)
    scrape = requests.get(url_page)
    soup = BeautifulSoup(scrape.content, 'lxml')
    link = soup.find_all('div',{'class':'item','class':'summary'})
    length = len(link)
    for i in range(0,length):
        name = link[i].a.get('title')
        product_name.append(name)
```

```python
# write to csv
# convert the list to a pandas dataframe

df = pd.DataFrame({'name':product_name})
df
df.to_csv('output.csv', index=False)
```

Scrape from Car List

```python
url1 = 'https://www.carlist.my/cars-for-sale/malaysia?hotdeal=true&page_number='
url2 = '&page_size=25'

price = [] #
discount = [] #
seller_type = []#
monthly_installment = []
product_brand = [] #
product_model =[] #
mileage = []
year = []

for page in range (1,13):
    url_page = url1+str(page)+url2
    scrape = requests.get(url_page)
    soup = BeautifulSoup(scrape.content, 'lxml')
    
    link = soup.find_all('article') 
    length = len(link)
    
    for i in range(0,length):
        price.append(link[i].get('data-hot-deal-actual-price'))
        discount.append(link[i].get('data-hot-deal-discount'))
        seller_type.append(link[i].get('data-ad-type'))
        
        #view = re.search(r'\d+,\d+',link[i].get('data-installment'))
        #view = view.group()
        #monthly_installment.append(view)
        
        product_brand.append(link[i].get('data-make'))
        product_model.append(link[i].get('data-model'))
        mileage.append(link[i].get('data-mileage'))
        year.append(link[i].get('data-year'))

print('Done!')        
        
df = pd.DataFrame({'brand':product_brand, 'model':product_model, 'type':seller_type, 'price':price, 'discount':discount, 'mileage':mileage, 'year':year})
df.head()
```

```python
df = df[df.price != '0']
df.price = df.price.str.replace("RM",'')
df.price = df.price.str.replace(",",'')

df.discount = df.discount.convert_objects(convert_numeric = True)
df.price = df.price.convert_objects(convert_numeric = True)
df['age'] = 2017-(df.year.convert_objects(convert_numeric = True))

df.info()
```

```python
# Brand with car sold

df.groupby('brand').size()
```

```python
import seaborn as sns

fig,axes = plt.subplots(figsize=(12,7))

brand_name = df.groupby('brand').size().index
brand = df.groupby('brand').size()
ind = [x for x, _ in enumerate(brand)]

axes.bar(ind, brand, align="center", width=0.5, alpha=0.5)
axes.set_xticks(ind)
axes.set_xticklabels(brand_name, fontsize=9, rotation = 'vertical')
axes.set_title("Number of Car Sales by Car Brand")
```

```python
fig,axes = plt.subplots(figsize=(12,7))

age_name = df.groupby('age').size().index
age = df.groupby('age').size()
ind = [x for x, _ in enumerate(age)]

axes.bar(ind, age, align="center", width=0.5, alpha=0.5)
axes.set_xticks(ind)
axes.set_xticklabels(age_name, fontsize=9, rotation = 'vertical')
axes.set_title("Number of Car Sales by Car Age")
```

```python
fig,axes = plt.subplots(figsize=(12,7))

brand_name = df.groupby('brand').size().index
brand = df.groupby('age').mean()['discount']
ind = [x for x, _ in enumerate(brand)]

axes.bar(ind, brand, align="center", width=0.5, alpha=0.5)
axes.set_xticks(ind)
axes.set_xticklabels(brand_name, fontsize=9, rotation = 'vertical')
axes.set_title("Average Discount by Car Brand")
```

```python
fig,axes = plt.subplots(1,2, figsize=(12,5))

age1 = df[df.brand == 'Toyota'].groupby('age').mean().index
price1 = df[df.brand == 'Toyota'].groupby('age').mean()['price']
ind1 = [x for x, _ in enumerate(age1)]

axes[0].plot(ind1,price1, color='#FF7F7F', label='Price')
axes[0].legend(loc=1)
axes[0].set_xticks(ind1)
axes[0].set_xticklabels(age1, fontsize=9, rotation = 'vertical')

axes[0].set_xlabel('Car Age')
axes[0].set_ylabel('Car Price, RM')
axes[0].set_title ('Toyota')

age2 = df[df.brand == 'Nissan'].groupby('age').mean().index
price2 = df[df.brand == 'Nissan'].groupby('age').mean()['price']
ind2 = [x for x, _ in enumerate(age2)]

axes[1].plot(ind2,price2, color='#FF7F7F', label='Price')
axes[1].legend(loc=1)
axes[1].set_xticks(ind2)
axes[1].set_xticklabels(age2, fontsize=9, rotation = 'vertical')

axes[1].set_xlabel('Car Age')
axes[1].set_ylabel('Car Price, RM')
axes[1].set_title ('Nissan')

fig.tight_layout()
```
